```python
import pandas as pd
import time
from bs4 import BeautifulSoup
from urllib.request import urlopen
from openpyxl import Workbook
from openpyxl.styles import Alignment, Font
wb = Workbook()
file = wb.active
file['A1'] = "Name".upper()
file['B1'] = 'Specifications'.upper()
file['C1'] = 'Price'.upper()
file['A1'].font = Font(bold=True)
file['B1'].font = Font(bold=True)
file['C1'].font = Font(bold=True)
file.column_dimensions['A'].width = 40
file.column_dimensions['b'].width = 40
file.column_dimensions['C'].width = 20
names = []
specs_list = []
price_list = []
d=''
wrap_alignment = Alignment(wrapText=True)

product = input("Search smartphone:")
product = product.replace(" ","%20")
pages = int(input("Number of pages to extract:"))
for i in range(1,pages+1):
    url=f"https://www.flipkart.com/search?q={product}&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"
    html = urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser')
    for ext in soup.find_all('div',class_='_2kHMtA'):
        mobile_name = ext.find('div',class_='col col-7-12').div.text
        print(mobile_name)
        names.append(mobile_name)
        print(" ")
        all_specs = ext.find('div',class_='col col-7-12').ul
        specs = all_specs.find_all('li')
        for i in  range(0,len(specs)):
            d = d + specs[i].text + " " 
        specs_list.append(d)
        d=''
        print("Specs:")
        for i in range(0,len(specs)):
            print(specs[i].text)
        all_price =  ext.find('div',class_='_3tbKJL').div
        price = all_price.find_all('div')
        print(" ")
        print(f"Current price:{price[0].text}")
        price_list.append(price[0].text)
        print(' ')
        print(' ')
for i in range(1,len(names)+1):
    file.cell(row=i+1,column = 1,value = names[i-1])
    file[f'A{i+1}'].alignment = Alignment(horizontal="center",vertical="center")
    file.cell(row=i+1,column = 2,value = specs_list[i-1])
    file[f'B{i+1}'].alignment = Alignment(wrap_text=True)
    file.cell(row=i+1,column = 3,value = price_list[i-1])
    file[f'C{i+1}'].alignment = Alignment(horizontal="center",vertical="center")
wb.save("Output.xlsx")
```
